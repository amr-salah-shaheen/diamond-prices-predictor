# Diamond Price Prediction
---

## 1. PROBLEM DESCRIPTION

### Problem Statement
Given a set of physical and quality attributes of a diamond, predict its **price in US dollars**. This is a supervised regression problem.

### Goal
Build a robust, production-ready ML pipeline that accurately predicts diamond prices based on measurable characteristics, enabling jewelers, buyers, and analysts to make data-driven pricing decisions.

### Features Description

| Feature | Type | Description |
|---------|------|-------------|
| `carat` | Numerical (float) | Weight of the diamond |
| `cut` | Categorical (ordinal) | Quality of the cut: Fair < Good < Very Good < Premium < Ideal |
| `color` | Categorical (ordinal) | Diamond color grade: D (best) to J (worst) |
| `clarity` | Categorical (ordinal) | Clarity grade: I1 < SI2 < SI1 < VS2 < VS1 < VVS2 < VVS1 < IF |
| `depth` | Numerical (float) | Total depth percentage: z / mean(x, y) × 100 |
| `table` | Numerical (float) | Width of top of diamond relative to widest point |
| `x` | Numerical (float) | Length in mm |
| `y` | Numerical (float) | Width in mm |
| `z` | Numerical (float) | Depth in mm |

### Target Variable
- **`price`** — Price in US dollars (\$326–\$18,823)

### Metrics
| Metric | Description |
|--------|-------------|
| **RMSE** | penalises large errors |
| **MAE** | interpretable dollar error |
| **R²** | variance explained |

## 4. DATA PREPROCESSING

Handle duplicates, physically impossible values (zero dimensions), and outliers using the IQR method.

```python
# ── Remove Duplicates ──
df = df.drop_duplicates()
print(f"Rows remaining: {len(df):,}")
```

```python
# ── Remove Physically Impossible Values ──
# Diamonds with x, y, or z == 0 are physically impossible — remove them
zero_dimension_rows = (df['x'] == 0) | (df['y'] == 0) | (df['z'] == 0)
print(f"Rows with zero dimensions: {zero_dimension_rows.sum()}")
df = df[~zero_dimension_rows].reset_index(drop=True)
print(f"Rows after removal: {len(df):,}")
```

```python
# ── 4.3 Outlier Treatment — IQR Capping ──

NUMERICAL_FEATURES = ['carat', 'depth', 'table', 'x', 'y', 'z']
IQR_FACTOR = 3
bounds = {}
for col in NUMERICAL_FEATURES:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - IQR_FACTOR * IQR
    upper = Q3 + IQR_FACTOR * IQR
    bounds[col] = (lower, upper)

outliers_mask = pd.Series(False, index=df.index)
for col in NUMERICAL_FEATURES:
    lower, upper = bounds[col]
    outliers_mask |= (df[col] < lower) | (df[col] > upper)

pct_outliers = outliers_mask.sum() / len(df) * 100
print(f"rows that will be capped : {outliers_mask.sum():,} ({pct_outliers:.1f}%)")

outlier_summary = []
for col in NUMERICAL_FEATURES:
    lower, upper = bounds[col]
    n_out = ((df[col] < lower) | (df[col] > upper)).sum()
    df[col] = df[col].clip(lower, upper)
    outlier_summary.append({'feature': col,'n_capped': n_out})

pd.DataFrame(outlier_summary)
```

I chose an IQR factor of 3 because the dataset contains large values that are valid and meaningful, not errors. Using a smaller factor like 1.5 would risk capping or removing important observations, so 3 helps preserve the true data distribution while still handling extreme cases

### Preprocessing Summary
- **Duplicates**: Removed any identical rows found
- **Zero-dimension rows**: Removed physically impossible entries (x/y/z = 0)
- **Outliers**: Applied 3×IQR Winsorisation on numerical features
- **No encoding or scaling here** — handled inside each model pipeline to prevent leakage

## 6. FEATURE ENGINEERING
Create informative new features and eliminate redundancy.

```python
df_fe = df.copy()

# ── Volume Feature ──
# x, y, z are individually highly correlated with each other and with carat.
# Combining them into a single 'volume' reduces multicollinearity.
df_fe['volume'] = df_fe['x'] * df_fe['y'] * df_fe['z']
print("Created: volume = x * y * z")
```

```python
# ── Drop Redundant Features ──
# x, y, z are now captured by `volume` and are redundant.
# Dropping them reduces dimensionality and multicollinearity.
COLS_TO_DROP = ['x', 'y', 'z']
df_fe = df_fe.drop(columns=COLS_TO_DROP)
print(f"Dropped columns: {COLS_TO_DROP}")
print(f"Remaining columns: {list(df_fe.columns)}")
```

```python
# Final feature set
FEATURE_COLS = ['carat', 'cut', 'color', 'clarity','depth', 'table', 'volume']
print(f"Final features ({len(FEATURE_COLS)}): {FEATURE_COLS}")
print(f"Target: {TARGET}")
```

**Feature Engineering Summary:**
- `volume = x*y*z` — single size metric replacing 3 correlated columns
- Dropped `x`, `y`, `z` — redundant given volume

## 8. Preprocessing Pipeline Design & Evaluation Function

Two preprocessing pipelines are defined based on each model's requirements:

| Pipeline | Encoding | Scaling | Models |
| --- | --- | --- | --- |
| ``preprocessor_scaled`` | ✅ OrdinalEncoder | ✅ RobustScaler | LinearRegression, KNN|
| ``preprocessor_no_scale`` | ✅ OrdinalEncoder | ❌ None | DecisionTree, RandomForest, GradientBoosting, XGBoost, LightGBM |

```python
NUMERICAL_COLS   = ['carat', 'depth', 'table', 'volume']
CATEGORICAL_COLS = ['cut', 'color', 'clarity']

# OrdinalEncoder — preserves grade hierarchy for both pipelines
from sklearn.base import clone
ordinal_enc_scaled   = OrdinalEncoder(categories=[CUT_ORDER, COLOR_ORDER, CLARITY_ORDER],
                                      handle_unknown='use_encoded_value', unknown_value=-1)

ordinal_enc_noscale  = clone(ordinal_enc_scaled)  # independent copy

# ── Pipeline A: with scaling ──
# RobustScaler: median-based — robust to outliers
preprocessor_scaled = ColumnTransformer([
    ('ord', ordinal_enc_scaled, CATEGORICAL_COLS),
    ('num', RobustScaler(), NUMERICAL_COLS)
])

# ── Pipeline B: without scaling ──
# Trees only need ordinal encoding — scaling has zero effect on splits
preprocessor_no_scale = ColumnTransformer([
    ('ord', ordinal_enc_noscale, CATEGORICAL_COLS),
    ('num', 'passthrough', NUMERICAL_COLS)
])
```

```python
def model_cv_evaluation(pipeline, X_tr, y_tr, name, cv=CV_FOLDS):
    scores = cross_validate(
        pipeline,
        X_tr,
        y_tr,
        cv=cv,
        scoring={
            "rmse": "neg_root_mean_squared_error",
            "mae":  "neg_mean_absolute_error",
            "r2":   "r2",
        },
        return_train_score=True,
        n_jobs=-1,
    )
    
    cv_rmse  = -scores["test_rmse"].mean()
    cv_mae   = -scores["test_mae"].mean()
    cv_r2    =  scores["test_r2"].mean()
    tr_rmse  = -scores["train_rmse"].mean()
    tr_mae   = -scores["train_mae"].mean()
    tr_r2    =  scores["train_r2"].mean()
    
    print(f"{'':6s}  {'RMSE':>8}  {'MAE':>7}  {'R²':>7}")
    print(f"  Train  {tr_rmse:8.1f}  {tr_mae:7.1f}  {tr_r2:7.4f}")
    print(f"  CV     {cv_rmse:8.1f}  {cv_mae:7.1f}  {cv_r2:7.4f}")
    return {
        "model":      name,
        "train_RMSE": tr_rmse,
        "cv_RMSE":    cv_rmse,
        "train_MAE":  tr_mae,
        "cv_MAE":     cv_mae,
        "train_R2":   tr_r2,
        "cv_R2":      cv_r2,
        "overfit":    tr_r2 - cv_r2,
    }
    
models_results = []
```

---
### 9. Model Training

#### Model 1 — Linear Regression

```python
pipe_lr = Pipeline([('pre', preprocessor_scaled),
                    ('model', LinearRegression())])
print('Linear Regression')
r_lr = model_cv_evaluation(pipe_lr, X_train, y_train, "LinearRegression")
models_results.append(r_lr)
```

#### Model 2 — KNN Regressor

```python
pipe_knn = Pipeline([('pre', preprocessor_scaled),
                     ('model', KNeighborsRegressor(n_jobs=-1))])
print('KNN Regressor')
r_knn = model_cv_evaluation(pipe_knn, X_train, y_train, name="KNN")
models_results.append(r_knn)
```

#### Model 3 — Decision Tree

```python
pipe_dt = Pipeline([('pre', preprocessor_no_scale),
                    ('model', DecisionTreeRegressor())])
print('Decision Tree')
r_dt = model_cv_evaluation(pipe_dt, X_train, y_train, name="DecisionTree")
models_results.append(r_dt)
```

#### Model 4 — Random Forest

```python
pipe_rf = Pipeline([('pre', preprocessor_no_scale),
                    ('model', RandomForestRegressor(n_jobs=-1, random_state=RANDOM_STATE))])
print('Random Forest')
r_rf = model_cv_evaluation(pipe_rf, X_train, y_train, name="RandomForest")
models_results.append(r_rf)
```

#### Model 5 — Gradient Boosting

```python
pipe_gb = Pipeline([('pre', preprocessor_no_scale),
                    ('model', GradientBoostingRegressor(random_state=RANDOM_STATE))])
print('Gradient Boosting')
r_gb = model_cv_evaluation(pipe_gb, X_train, y_train, name="GradientBoosting")
models_results.append(r_gb)
```

#### Model 6 — XGBoost

```python
pipe_xgb = Pipeline([('pre', preprocessor_no_scale),
                     ('model', XGBRegressor(random_state=RANDOM_STATE, n_jobs=-1))])
print('XGBoost')
r_xgb = model_cv_evaluation(pipe_xgb, X_train, y_train, name="XGBoost")
models_results.append(r_xgb)
```

#### Model 7 — LightGBM

```python
pipe_lgbm = Pipeline([('pre', preprocessor_no_scale),
                      ('model', LGBMRegressor(random_state=RANDOM_STATE, n_jobs=-1,verbose=-1))])
print('LightGBM')
r_lgbm = model_cv_evaluation(pipe_lgbm, X_train, y_train, name="LightGBM")
models_results.append(r_lgbm)
```

#### Baseline Summary
Compare all models side-by-side — train vs validation to spot overfitting.

```python
# ── Baseline Comparison Table ──
summary = pd.DataFrame([
    {
        "Model":      r["model"],
        "Train RMSE": round(r["train_RMSE"], 1),
        "CV RMSE":    round(r["cv_RMSE"],    1),
        "Train MAE":  round(r["train_MAE"],  1),
        "CV MAE":     round(r["cv_MAE"],     1),
        "Train R²":   round(r["train_R2"],   4),
        "CV R²":      round(r["cv_R2"],      4),
        "Overfit":    round(r["overfit"],    4),
    }
    for r in models_results
]).sort_values("CV R²", ascending=False)

df_summary = pd.DataFrame(summary)
df_summary
```

## 10. HYPERPARAMETER TUNING

GridSearchCV with 5-fold cross-validation on the top 3 performers from baseline.

```python
# ── 9.1 Random Forest Tuning ──
rf_param_dist = {
    'model__n_estimators':      [200, 300, 400],
    'model__max_depth':         [10, 20, 30],
    'model__min_samples_split': [2, 5, 10],
    'model__max_features':      [None,'sqrt', 0.5]
}

rf_random_search = RandomizedSearchCV(
    estimator=pipe_rf,
    param_distributions=rf_param_dist,
    n_iter=50,
    cv=CV_FOLDS,
    n_jobs=-1,
    verbose=1,
    random_state=RANDOM_STATE
)

print("Tuning RandomForest with RandomizedSearch ...")
rf_random_search.fit(X_train, y_train)
print(f"  Best params : {rf_random_search.best_params_}")
print(f"  Best CV R²  : {rf_random_search.best_score_:.4f}")
pd.DataFrame(rf_random_search.cv_results_).sort_values('std_test_score')
```

```python
# ── 9.2 XGBoost Tuning ──
xgb_param_grid = {
    'model__n_estimators':      [200, 300, 400],
    'model__learning_rate':     [0.01, 0.05, 0.1],
    'model__max_depth':         [4, 6, 8],
    'model__subsample':          [0.6, 0.8, 1.0],
    'model__colsample_bytree':   [0.6, 0.8, 1.0],
}
xgb_random_search = RandomizedSearchCV(
    pipe_xgb,
    param_distributions=xgb_param_grid,
    n_iter=50,
    cv=CV_FOLDS,
    n_jobs=-1,
    verbose=1,
    random_state=RANDOM_STATE
)
print('Tuning XGBoost ...')
xgb_random_search.fit(X_train, y_train)
print(f'  Best params : {xgb_random_search.best_params_}')
print(f'  Best CV R²  : {xgb_random_search.best_score_:.4f}')
pd.DataFrame(xgb_random_search.cv_results_).sort_values('std_test_score')
```

```python
# ── 9.3 LightGBM Tuning ──
lgbm_param_grid = {
    'model__n_estimators':  [100, 200, 300],
    'model__learning_rate': [0.01, 0.05, 0.1],
    'model__max_depth':     [5, 10, 20],
    'model__num_leaves':    [31, 63,127],
    'model__subsample':     [0.6, 0.8, 1.0]
}
lgbm_random_search = RandomizedSearchCV(
    pipe_lgbm,
    param_distributions=lgbm_param_grid,
    n_iter=50,
    cv=CV_FOLDS,
    n_jobs=-1,
    verbose=1,
    random_state=RANDOM_STATE
)
print('Tuning LightGBM ...')
lgbm_random_search.fit(X_train, y_train)
print(f'  Best params : {lgbm_random_search.best_params_}')
print(f'  Best CV R²  : {lgbm_random_search.best_score_:.4f}')
pd.DataFrame(lgbm_random_search.cv_results_).sort_values('std_test_score')
```

### Choosing the best model

```python
tuned_candidates = {
    "RandomForest": rf_random_search.best_estimator_,
    "XGBoost":      xgb_random_search.best_estimator_,
    "LightGBM":     lgbm_random_search.best_estimator_,
}

print("Tuned model comparison on X_val (first and only val-set contact):")
print(f"{'Model':>13}  {'Val R²':>10} ")

best_name, best_model, best_val_r2 = None, None, -np.inf
for name, est in tuned_candidates.items():
    val_r2   = est.score(X_val, y_val)
    val_pred = est.predict(X_val)
    print(f"{name:>15}  {val_r2:8.4f}")
    if val_r2 > best_val_r2:
        best_val_r2  = val_r2
        best_name    = name
        best_model   = est

print(f"\n Best tuned model: {best_name}  (Val R² = {best_val_r2:.4f})")
```

## 11. MODEL EVALUATION

```python
# ── Evaluate Best Model on Test Set ──
y_pred_test = best_model.predict(X_test)
test_rmse = root_mean_squared_error(y_test, y_pred_test)
test_mae  = mean_absolute_error(y_test, y_pred_test)
test_r2   = r2_score(y_test, y_pred_test)

print(f'Test Set Results — {best_name} (tuned)')
print('=' * 40)
print(f'  RMSE : {test_rmse:,.2f}')
print(f'  MAE  : {test_mae:,.2f}')
print(f'  R²   : {test_r2:.4f}')
print('=' * 40)
```

```python
# ── Residual Analysis — Best Model on Test Set ──

residuals = y_test - y_pred_test
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Predicted vs Actual
axes[0].scatter(y_test, y_pred_test, alpha=0.2, s=5, color='steelblue')
lims = [min(y_test.min(), y_pred_test.min()), max(y_test.max(), y_pred_test.max())]
axes[0].plot(lims, lims, 'r--', linewidth=1.5)
axes[0].set_xlabel('Actual Price')
axes[0].set_ylabel('Predicted Price')
axes[0].set_title('Predicted vs Actual', fontweight='bold')

# Residuals vs Predicted
axes[1].scatter(y_pred_test, residuals, alpha=0.2, s=5, color='coral')
axes[1].axhline(0, color='k', linewidth=1.2)
axes[1].set_xlabel('Predicted Price')
axes[1].set_ylabel('Residual')
axes[1].set_title('Residuals vs Predicted', fontweight='bold')

# Residual distribution
sns.histplot(residuals, kde=True, ax=axes[2], color='lightgreen')
axes[2].set_title('Residual Distribution', fontweight='bold')

plt.suptitle(f'Residual Analysis — {best_name} (tuned) on Test Set', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
